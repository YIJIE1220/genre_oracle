```python
from typing import List, Dict
import pandas as pd
import json
from pandas.io.json import json_normalize
import os
```

```python

data = json.load(open('spotify.json'))
df=json_normalize(data,'artists',['popularity','duration_ms','id','name'],record_prefix='_')
df_spotify=df[['_id','_name','popularity','id','duration_ms','name']]
df_spotify.columns = ['main_artist_id', 'main_artist_name', 'popularity', 'song_id','song_length','song_name']
df_spotify['main_artist_name']=df_spotify['main_artist_name'].str.lower()
df_spotify['song_name']=df_spotify['song_name'].str.lower()

print(df_spotify.head())
```

```python
print('old shape:',df_spotify.shape)
df_spotify.dropna
print('new shape:',df_spotify.shape)
```

```python
df_spotify=df_spotify.drop_duplicates('song_id')
```

```python
df_spotify=df_spotify.sort_values(by =['main_artist_name','song_name','popularity'],ascending=False)
df_spotify
```

```python
df_spotify=df_spotify.drop_duplicates(subset=['main_artist_name','song_name'])
print('new shape:',df_spotify.shape)
```

```python
df_spotify=df_spotify.sort_values(by =['popularity'],ascending=False)
df_spotify
```
