```python
import pandas as pd
lyrics = pd.read_csv("lyrics_cleaned_desc_by_count.csv")
songs = pd.read_csv("merged_dropdup.csv")

"""https://www.noswearing.com"""
#list of all known profaninity and slang! Used web scraper from this guy: http://tylermarrs.com/tag/python/
profanity = pd.read_csv("profanity.txt", names = ["word"])
profanity = profanity.sort_values('word', ascending = True)
```

```python
"""create word count metric!"""
hold_count = lyrics.groupby("track_id")["count"].agg("sum").reset_index()
songs = pd.merge(songs, hold_count, on='track_id')
songs = songs.rename(columns={"count":"word_count"})
print(len(songs))

"""create bool word: i.e., if it is profanity or not"""
profarr = list(profanity.word.apply(lambda x: x))
lyrics["prof_bool"] = ''
lyrics.prof_bool = lyrics.word.apply(lambda x: x in profarr)
```

```python
hold_prof_count = lyrics.groupby("track_id")["prof_bool"].agg("sum").reset_index()

#Remove all negatives and then perform!
lyrics = lyrics[lyrics.prof_bool == True]

#get the profanity count
hold_prof_count = lyrics.groupby("track_id")["count"].agg("sum").reset_index()

#songs["prof_count"] = songs.track_id.apply(lambda x: lyrics.loc[lyrics.track_id == x]["count"].sum())

#merge
songs = pd.merge(songs, hold_prof_count, how = 'outer', on="track_id")
songs = songs.rename(columns={"count":"prof_count"})


#outer merge imputes nan values; iff missing outer, will only merge matching entries! 
```

```python
#replace the nan values with 0 as that mean they did not match the merge true conditionel
songs.prof_count = songs.prof_count.fillna(0)
```

```python
#create new column for prof_percentage
songs["prof_percen"] = songs["prof_count"] / songs["word_count"] 
```

```python
#gp == genre profanity!!! Profanity by genre
gp = pd.DataFrame()
gp["genre_merged"] = songs["genre_merged"].unique()

#average perc and count per genre
hold_percen = songs.groupby("genre_merged")["prof_percen"].agg("mean").reset_index()
hold_avg_prof_count = songs.groupby("genre_merged")["prof_count"].agg("mean").reset_index()
gp = pd.merge(gp, hold_percen, on="genre_merged")
gp = pd.merge(gp, hold_avg_prof_count, on="genre_merged")
gp = gp.rename(columns={"genre_merged":"genre", "prof_percen":"avg_prof_percen"
                       ,"prof_count":"avg_prof_wc"})
gp
```

```python
#sub to csvs!!!!
gp.to_csv('genre_profanity.csv')
songs.to_csv("songs_merged_profanity.csv")


```
